# Task 1B: KM trên Ụ và sau thu ngân, cái nào hiệu quả hơn?

```python
master_df['Trưng bày'] = master_df['Trưng bày'].str.replace('Trên Kệ','Trên kệ', regex =True)
```

```python
master_df['Trưng bày'].unique()
```

```python
df_1b = master_df[['Mã CTKM', 'CTKM', 'Thời gian khuyến mãi', 'Tỷ lệ khuyến mãi','Trưng bày','Mã SP chính']]
df_1b
```

```python
df_1b = df_1b[df_1b['Trưng bày'] != 'Trên kệ']
df_1b
```

```python
df_1b['Tổng doanh thu'] = master_df['Doanh thu KM'] + (master_df['Sức bán thường']* master_df['Giá bán KV chuẩn'])
df_1b['Tổng doanh thu'] = round(df_1b['Tổng doanh thu'], 2)
df_1b
```

```python
df_1b['Khoảng TGKM'] = '0'
df_1b['Khoảng TGKM'].loc[(df_1b['Thời gian khuyến mãi'] <= 7)] = '<= 7'
df_1b['Khoảng TGKM'].loc[(df_1b['Thời gian khuyến mãi'] > 7) & (df_1b['Thời gian khuyến mãi'] <= 14)] = '8-14'
df_1b['Khoảng TGKM'].loc[(df_1b['Thời gian khuyến mãi'] > 14) & (df_1b['Thời gian khuyến mãi'] <= 24)] = '15-24'
df_1b['Khoảng TGKM'].loc[(df_1b['Thời gian khuyến mãi'] > 24) & (df_1b['Thời gian khuyến mãi'] <= 28)] = '25-28'
df_1b['Khoảng TGKM'].loc[(df_1b['Thời gian khuyến mãi'] > 28)] = '> 28'
df_1b
```

## Cơ cấu doanh thu theo cùng ngày KM:

```python
df_1b_cungngay = df_1b.groupby(['Trưng bày','Khoảng TGKM'])['Tổng doanh thu'].sum().reset_index().sort_values(['Khoảng TGKM'], ascending = False)
df_1b_cungngay['Trưng bày'] = df_1b_cungngay['Trưng bày'].sort_values(ascending = False)
df_1b_cungngay
```

```python
df_1b_cungngay = df_1b_cungngay[(df_1b_cungngay['Khoảng TGKM'] == '<= 7')| (df_1b_cungngay['Khoảng TGKM'] == '8-14')]
df_1b_cungngay
```

```python
khoangTGKM = df_1b_cungngay['Khoảng TGKM'].unique()
fig = plt.figure(figsize=(10, 10))
for i in range(len(khoangTGKM)):
    khoang = khoangTGKM[i]
    df_khoang = df_1b_cungngay[df_1b_cungngay['Khoảng TGKM'] == khoang]
    labels = df_khoang['Trưng bày'].unique()
    values = df_khoang.groupby('Trưng bày')['Tổng doanh thu'].sum()
    ax = fig.add_subplot(1, 2, i+1)
    ax.pie(values, labels=labels, autopct='%1.1f%%')
    ax.set_title('Khoảng TGKM: {}'.format(khoang))
    fig.add_subplot(ax)
plt.legend()
plt.show();
```

## Cơ cấu doanh thu theo cùng số sản phẩm:

```python
df_1b_cungsp = df_1b.groupby(['Trưng bày','Mã SP chính'])['Tổng doanh thu'].sum().reset_index()
df_1b_cungsp
```

```python
df_1b_cungsp = df_1b_cungsp.sort_values(by = 'Mã SP chính', ascending = True)
df_1b_cungsp
```

```python
grouped_df_1b_cungsp = df_1b_cungsp.groupby('Mã SP chính')
filtered_df = grouped_df_1b_cungsp.filter(lambda x: set(x['Trưng bày']) == set(['Tháp KM', 'Thu ngân']))
```

```python
filtered_df_sum1b = filtered_df.groupby('Trưng bày')['Tổng doanh thu'].sum().reset_index()
filtered_df_sum1b
```

```python
plt.figure(figsize=(6,6))
plt.title('Cơ cấu doanh thu cùng số sản phẩm theo trưng bày')
sns.set_palette('pastel')
sns.set_style('whitegrid')
plt.pie(filtered_df_sum1b['Tổng doanh thu'], labels=filtered_df_sum1b['Trưng bày'], autopct='%1.1f%%')
plt.legend()
plt.show();
```

## Tỷ lệ giảm giá trung bình giữa tháp KM và thu ngân với cùng số ngày KM:

```python
df_1b_cungngaykm = df_1b.groupby(['Trưng bày','Khoảng TGKM'])['Tỷ lệ khuyến mãi'].mean().reset_index().sort_values(['Khoảng TGKM'], ascending = False)
df_1b_cungngaykm['Trưng bày'] = df_1b_cungngaykm['Trưng bày'].sort_values(ascending = False)
df_1b_cungngaykm
```

```python
df_1b_cungngaykm = df_1b_cungngaykm[(df_1b_cungngaykm['Khoảng TGKM'] == '<= 7')| (df_1b_cungngaykm['Khoảng TGKM'] == '8-14')]
df_1b_cungngaykm
```

```python
khoangTGKM = df_1b_cungngaykm['Khoảng TGKM'].unique()
fig, axs = plt.subplots(nrows=1, ncols=len(khoangTGKM), figsize=(20, 5))

for i, khoang in enumerate(khoangTGKM):
    df_khoang = df_1b_cungngaykm[df_1b_cungngaykm['Khoảng TGKM'] == khoang]
    labels = df_khoang['Trưng bày'].unique()
    colors = ['blue', 'orange']
    values = df_khoang.groupby('Trưng bày')['Tỷ lệ khuyến mãi'].mean()
    ax = axs[i]
    ax.bar(labels, values, color=colors, width=0.3, align='center', alpha=1)
    ax.set_title('Khoảng TGKM: {}'.format(khoang))
    for j, v in enumerate(values):
        ax.text(j, v+0.05, '{:.1f}%'.format(v*100), ha='center')
    ax.set_ylim([0, 1])
plt.subplots_adjust(wspace=0.3)
plt.show();
```

## Cách 2: Cùng số CTKM Unique, count số lượng mẫu:

```python
df_new = df_1b.groupby(['Mã CTKM', 'CTKM', 'Khoảng TGKM', 'Trưng bày']).agg({'Mã SP chính': ['nunique', 'size'], 'Tổng doanh thu': 'sum', 'Tỷ lệ khuyến mãi': 'mean'})
df_new
```

```python
df_new.columns = ['_'.join(col) for col in df_new.columns]
df_new = df_new.reset_index().rename(columns={'Mã SP chính_nunique': 'Số lượng Mã SP chính khác nhau', 'Mã SP chính_size': 'Tổng số lượng Mã SP chính', 'Tổng doanh thu_sum': 'Tổng doanh thu', 'Tỷ lệ khuyến mãi_mean':'Tỷ lệ khuyến mãi'})
df_new = df_new.sort_values(by = 'Khoảng TGKM')
df_new
```

```python
df_unite = df_new.groupby(['Khoảng TGKM','Trưng bày']).agg({'Tổng số lượng Mã SP chính': 'sum', 'Tổng doanh thu': 'sum', 'Tỷ lệ khuyến mãi': 'mean'}).reset_index()
df_unite
```

```python
khoangTGKM = df_unite['Khoảng TGKM'].unique()
fig = plt.figure(figsize=(10, 10))
for i in range(len(khoangTGKM)):
    khoang = khoangTGKM[i]
    df_khoang = df_unite[df_unite['Khoảng TGKM'] == khoang]
    labels = df_khoang['Trưng bày'].unique()
    values = df_khoang.groupby('Trưng bày')['Tổng doanh thu'].sum()
    ax = fig.add_subplot(2, 2, i+1)
    ax.pie(values, labels=labels, autopct='%1.1f%%')
    ax.set_title('Khoảng TGKM: {}'.format(khoang))
    fig.add_subplot(ax)
plt.legend()
plt.show();
```

```python
df_unite['avg_doanhthu'] = df_unite['Tổng doanh thu'] / df_unite ['Tổng số lượng Mã SP chính']
df_unite
```

```python
khoangTGKM = df_unite['Khoảng TGKM'].unique()
fig = plt.figure(figsize=(10, 10))
for i in range(len(khoangTGKM)):
    khoang = khoangTGKM[i]
    df_khoang = df_unite[df_unite['Khoảng TGKM'] == khoang]
    labels = df_khoang['Trưng bày'].unique()
    values = df_khoang.groupby('Trưng bày')['avg_doanhthu'].sum()
    ax = fig.add_subplot(2, 2, i+1)
    ax.pie(values, labels=labels, autopct='%1.1f%%')
    ax.set_title('Khoảng TGKM: {}'.format(khoang))
    fig.add_subplot(ax)
plt.legend()
plt.show();
```
