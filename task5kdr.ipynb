# Import Data 2023:

```python
master_df1 = pd.read_excel('data_doanhthu_key_2023.xlsx')
master_df1
```

```python
master_df1.info()
```

```python
df1 = master_df1[master_df1['Thời gian khuyến mãi'] <= 30]
df1['Thời gian khuyến mãi'].unique()
```

```python
df1['CTKM'].unique()
```

```python
df1 = df1[(df1['CTKM'] == 'Giảm giá') | (df1['CTKM'] == 'Mua 2 tặng 1')]
```

```python
df1 = df1[(df1['Là thương hiệu nổi tiếng hay không'] == 'Yes') & (df1['Phân loại giá'] == 'Trung (15K - 25K) NL')]
df1['Phân loại giá'].unique()
```

```python
df1 = df1[(df1['phân loại size'] != 'Nhỏ (<=100g)')]
df1['phân loại size'].unique()
```

```python
df1
```

```python
sum_df1 = df1['Tổng doanh thu'].sum()
sum_df1
```

```python
sum_master_df1 = master_df1['Tổng doanh thu'].sum()
sum_master_df1
```

```python
data = {'DataFrame': ['Doanh thu hiệu quả', 'Doanh thu tổng'],
        'Tổng doanh thu': [sum_df1, sum_master_df1]}
df = pd.DataFrame(data)
df
```

```python
plt.figure(figsize=(6,6))
plt.title('Tỷ lệ doanh thu hiệu quả 2023')
sns.set_palette('pastel')
sns.set_style('whitegrid')
plt.pie(df['Tổng doanh thu'], labels=df['DataFrame'], autopct='%1.1f%%')
plt.legend()
plt.show()
```
