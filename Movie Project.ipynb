```python
df['movie_release_date'] = pd.to_datetime(df['movie_release_date'])
df.info()
```

```python
df['movie_release_year'] = df['movie_release_date'].dt.year
df['movie_release_year'] 
```

```python
df.columns
```

```python
df_movie = df[['movie_id', 'movie_title', 'movie_popularity', 'movie_release_date','movie_budget', 'movie_revenue','movie_release_year']].drop_duplicates()
df_movie.head()
```

```python
#Solution 1:
df_movie.nlargest(n = 10, columns = ['movie_revenue'])
```

```python
#Solution 2:
df_re  = df_movie.sort_values(by = 'movie_revenue', ascending = False)
df_re.head(10)
```

```python
df_gr = df_movie.groupby('movie_release_year')['movie_id'].nunique()
df_gr
```

```python
df_sum = df_movie.groupby('movie_release_year')['movie_revenue'].sum()
df_sum
```

```python
df_sum.sort_values(ascending = False)
```

```python
years = []
for year, revenue in df_sum.items():
    years.append(year)
    
plt.bar(x=years, height= df_sum, color = 'g')
plt.title('Best Year For Revenue', fontsize = 16)
plt.xticks(years,rotation = 90, fontsize = 12)
plt.xlabel('Years', fontsize = 12)
plt.ylabel('Revenue',fontsize = 12)

plt.show()
```

```python
sns.relplot(data=df_movie, x='movie_budget', y = 'movie_revenue');
```
