# Data Virtualization:

```python
df_corr_matrix = df.corr()
fig, ax = plt.subplots(figsize=(10, 10))
ax = sns.heatmap(df_corr_matrix, annot = True, cmap="Blues");
```

```python
sns.scatterplot(x='reading score', y='math score', data=df, hue='test preparation course', palette='coolwarm')
plt.xlim(0,105)
plt.ylim(0,105)
plt.show();
```

```python
sns.scatterplot(x='reading score',y='writing score',data=df,hue='test preparation course',palette='coolwarm')
plt.xlim(0,105)
plt.ylim(0,105)
plt.show();
```

```python
sns.pairplot(df, hue="test preparation course");
```

```python
pd.crosstab(df["race/ethnicity"], df.gender).plot(kind="bar", figsize=(10,6), color=["blue", "lightblue"]);
plt.title("Genders by Group")
plt.xlabel("0 = Group 1, 1 = Group B, 2 = Group C , 3 = Group D, 4 = Group E ")
plt.ylabel("Amount")
plt.legend(["Female", "Male"])
plt.show();
```
