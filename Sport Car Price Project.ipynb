# Import Library:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print('Setup Completed!')
```

# Import Data:

```python
df = pd.read_csv('Sport car price.csv')
df
```

```python
df.info()
```

# Data Cleaning:

```python
df.isnull().sum()
```

```python
df.duplicated().any()
```

```python
df[df.duplicated()]
```

```python
df = df.drop_duplicates()
df
```

```python
df = df.fillna('Not Available')
df
```

```python
df['Price (in USD)'] = df['Price (in USD)'].replace(',','', regex = True)
df['Price (in USD)'] = df['Price (in USD)'].astype('int')
df
```

# Data Virtualization:

## Top 10 Car Make Price:

```python
top10carmake_price = df.groupby('Car Make')['Price (in USD)'].sum().nlargest(10)
top10carmake_price
```

```python
df_top10carmake_price = top10carmake_price.to_frame('Price').rename_axis('Car Make').reset_index()
df_top10carmake_price
```

```python
plt.figure(figsize = (12,10))
ax = sns.barplot(x = df_top10carmake_price['Car Make'], y = df_top10carmake_price['Price'], order = top10carmake_price.index)
plt.xticks(rotation = 90)
plt.title('Top 10 Car Make Price')
plt.show();
```

## Top 10 Car Model Price:

```python
top10carmodel_price = df.groupby('Car Model')['Price (in USD)'].sum().nlargest(10)
top10carmodel_price
```

```python
df_top10carmodel_price = top10carmodel_price.to_frame('Price').rename_axis('Car Model').reset_index()
df_top10carmodel_price
```

```python
plt.figure(figsize = (12,10))
ax = sns.barplot(x = df_top10carmodel_price['Car Model'], y = df_top10carmodel_price['Price'], order = top10carmodel_price.index)
plt.xticks(rotation = 90)
plt.title('Top 10 Car Model Price')
plt.show();
```

## Top 10 Car Model Quantity:

```python
top10_quant_car = df.groupby('Car Make')['Car Model'].count().nlargest(10)
top10_quant_car
```

```python
df_top10_quant_car = top10_quant_car.to_frame('Quantity').rename_axis('Car Make').reset_index()
df_top10_quant_car
```

```python
plt.figure(figsize = (12,10))
ax = sns.barplot(x = df_top10_quant_car['Car Make'], y = df_top10_quant_car['Quantity'], order = top10_quant_car.index)
plt.xticks(rotation = 90)
plt.title('Top 10 Car Model Quantity')
plt.show();
```

## About Horsepower:

```python
df['Horsepower'].unique()
```

```python
df['Horsepower'] = df['Horsepower'].replace(',','', regex = True)
df['Horsepower'] = df['Horsepower'].replace('1000+','1000', regex = True)
df['Horsepower'] = df['Horsepower'].replace('10000+','10000', regex = True)
```

```python
df_horse = df[df['Horsepower'] != '1000+']
df_horse
```

```python
df_horse['Horsepower'] = df_horse['Horsepower'].astype('int')
df_horse.info()
```

```python
df_horse = df_horse[['Horsepower','Price (in USD)']]
df_horse
```

```python
plt.figure(figsize = (12,10))
ax = sns.jointplot(x = df_horse['Horsepower'], y = df_horse['Price (in USD)'])
plt.xticks(rotation = 90)
plt.show();
```
