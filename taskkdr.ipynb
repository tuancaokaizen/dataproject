# Import Library:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.subplots as ps
import plotly.graph_objects as go
from plotly.offline import plot

print('Setup Completed!')
```

# Import Data:

```python
df = pd.read_excel('data_2023_key.xlsx', sheet_name= 'Sheet1')
df.head()
```

```python
df.info()
```

```python
df['Doanh thu KM'] = df['promoquantity2'] * (df['Giá bán KV chuẩn'] *(1 - df['Tỷ lệ khuyến mãi']))
df.head()
```

```python
df['Doanh thu thường'] = df['Sức bán thường'] * df['Giá bán KV chuẩn']
df
```

```python
df['Tổng doanh thu'] =  df['Doanh thu KM'] + df['Doanh thu thường']
df
```

```python
df.to_excel('data_doanhthu_key_2023.xlsx')
```

# Trưng bày theo cụm:

## Phân tích số lượng trưng bày theo cụm:

```python
df_trungbay = df[['Cụm nhỏ','Trưng bày','Doanh thu KM', 'Doanh thu thường', 'Tổng doanh thu']]
df_trungbay
```

```python
df_trungbay['Trưng bày'].unique()
```

```python
df_trungbay['Trưng bày'] = df_trungbay['Trưng bày'].str.replace('Trên Kệ','Trên kệ', regex =True)
```

```python
count_trungbay = df_trungbay.groupby(['Cụm nhỏ','Trưng bày'])['Trưng bày'].count()
count_trungbay
```

```python
df_count_trungbay = count_trungbay.reset_index(name = 'count')
```

```python
df_count_trungbay['total'] = df_count_trungbay.groupby('Cụm nhỏ')['count'].transform('sum')
df_count_trungbay
```

```python
df_count_trungbay['percentage'] = df_count_trungbay['count']/df_count_trungbay['total']* 100
df_count_trungbay = df_count_trungbay.sort_values(by= 'Trưng bày', ascending = False)
df_count_trungbay
```

```python
ax = sns.catplot(x='Cụm nhỏ', y='count', hue='Trưng bày', data=df_count_trungbay, kind='bar', height=6, aspect=2)
plt.title('Số lượng trưng bày trong từng cụm nhỏ')
plt.xlabel('Cụm nhỏ')
plt.ylabel('Số lượng')
plt.show();
```

```python
fig, axs = plt.subplots(nrows=3, ncols=3, figsize=(10, 10))

for i, ax in enumerate(axs.flatten()):
    df_cum_nho = df_count_trungbay[df_count_trungbay['Cụm nhỏ'] == f'Cụm {i}']
    ax.pie(df_cum_nho['count'], labels=df_cum_nho['Trưng bày'], autopct='%1.1f%%', startangle=90)
    ax.set_title(f'Cụm {i}')

plt.tight_layout()
plt.legend()
plt.show();
```

## Phân tích theo doanh thu của trưng bày theo cụm:

```python
doanhthu_trungbay = df_trungbay.groupby(['Cụm nhỏ','Trưng bày'])['Tổng doanh thu'].sum()
doanhthu_trungbay
```

```python
df_doanhthu_trungbay = doanhthu_trungbay.reset_index(name = 'Tổng doanh thu')
df_doanhthu_trungbay
```

```python
df_doanhthu_trungbay['total'] = df_doanhthu_trungbay.groupby('Cụm nhỏ')['Tổng doanh thu'].transform('sum')
df_doanhthu_trungbay
```

```python
df_doanhthu_trungbay['percentage'] = df_doanhthu_trungbay['Tổng doanh thu']/df_doanhthu_trungbay['total']* 100
df_doanhthu_trungbay = df_doanhthu_trungbay.sort_values(by= 'Trưng bày', ascending = False)
df_doanhthu_trungbay
```

```python
plt.figure(figsize=(30, 30))
fig, axs = plt.subplots(nrows=3, ncols=3, figsize=(10, 10))

for i, ax in enumerate(axs.flatten()):
    df_cum_nho = df_doanhthu_trungbay[df_doanhthu_trungbay['Cụm nhỏ'] == f'Cụm {i}']
    ax.pie(df_cum_nho['Tổng doanh thu'], labels=df_cum_nho['Trưng bày'], autopct='%1.1f%%', startangle=90)
    ax.set_title(f'Cụm {i}')

plt.legend()
plt.tight_layout()
plt.show();
```
